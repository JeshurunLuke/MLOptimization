```python
import numpy as np
Translator = {'a': '11', 'b': '12', 'c': '13', 'd': '14', 'e': '15', 'f': '21', 'g': '22', 'h':'23', 'i':'24', 'j': '25',  'k': '31', 'l':'32', 'm': '33', 'n':'34', 'o':'35'}

def SeqTranslator(Seq): #Seq is a string
    key_list = list(Translator.keys())
    val_list = list(Translator.values())
    Array = []
    for pulse in range(0, len(Seq)-1, 2):
        pos  = val_list.index(Seq[pulse:pulse + 2])
        Array.append(key_list[pos])
    return ''.join(Array)

File_name = 'SeqCreaterSumSuccess.csv'    
df = pandas.read_csv(File_name)
for i in df['Seq']:
    finalSequence = SeqTranslator(i)


```

```python

print(f"{finalSequence.count('a')}: Axial -1")
print(f"{finalSequence.count('b')}: Axial -2")
print(f"{finalSequence.count('c')}: Axial -3")
print(f"{finalSequence.count('d')}: Axial -4")
print(f"{finalSequence.count('e')}: Axial -5")

print('')
print(f"{finalSequence.count('f')}: Radial_1 -1")
print(f"{finalSequence.count('g')}: Radial_1 -2")
print(f"{finalSequence.count('h')}: Radial_1 -3")
print(f"{finalSequence.count('i')}: Radial_1 -4")
print(f"{finalSequence.count('j')}: Radial_1 -5")


print('')
print(f"{finalSequence.count('k')}: Radial_2 -1")
print(f"{finalSequence.count('l')}: Radial_2 -2")
print(f"{finalSequence.count('m')}: Radial_2 -3")
print(f"{finalSequence.count('n')}: Radial_2 -4")
print(f"{finalSequence.count('o')}: Radial_2 -5")

```

```python
import matplotlib.pyplot as plt
axialData = []
radial1Data = []
radial2Data = []
axialCoord =['a','b', 'c', 'd', 'e']
radial1Coord = ['f', 'g', 'h', 'i', 'j']
radial2Coord = ['k', 'l', 'm', 'n', 'o']
for pos, i in enumerate(finalSequence):
    if i in axialCoord:
        axialData.append([pos, axialCoord.index(i)])
    elif i in radial1Coord:
        radial1Data.append([pos, radial1Coord.index(i)])
    elif i in radial2Coord:
        radial2Data.append([pos, radial2Coord.index(i)])
axialData = np.array(np.transpose(axialData), dtype = float)
radial1Data = np.transpose(radial1Data)
radial2Data = np.transpose(radial2Data)

plt.scatter(np.arange(len(axialData[1])), axialData[1])
plt.show()
plt.scatter(np.arange(len(radial1Data[1])), radial1Data[1])
plt.show()
plt.scatter(np.arange(len(radial2Data[1])), radial2Data[1])
plt.show()
print(len(axialData[0]), len(axialData[1]))
print(axialData[0])
plt.scatter(axialData[0], axialData[1])
plt.show()
plt.scatter(radial1Data[0], radial1Data[1], 'label = Radial 1')
plt.scatter(radial2Data[0], radial2Data[1],'label = Radial 2')
plt.show()
```
